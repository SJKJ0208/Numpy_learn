```python
from matplotlib import pyplot as plt 
import numpy as np
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

#设置x轴的刻度
#前面是使其密集，后者是步长使其稀疏
plt.xticks(np.arange(2,26,0.5)[::3])#直接将x值每个值都放入
plt.yticks(np.arange(min(y)-2,max(y)+2,0.5)[::3])#使其更加符合全图的尺寸

# 保存绘制后进行保存
plt.savefig("./i.svg")
plt.show()
```

```python
import random
x = np.arange(0,120)
a = [random.randint(20,35) for i in range(120)]
plt.figure(figsize = (20,8),dpi= 80)
_x = list(x)[::10]
#字符串加入到坐标轴中
_xtick_table = ["{} min ".format(i) for i in _x]

plt.plot(x,a)
#_x仅仅作为索引，真正起作用是第二个
plt.xticks(_x,_xtick_table)
plt.yticks(np.arange(min(a)-2,max(a)+2,1)[::1])


plt.show()
```
